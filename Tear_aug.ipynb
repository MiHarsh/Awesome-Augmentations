```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
```

```python
a = np.arange(256,0,-5)
y1 = np.random.randint(0,50)
y2 = np.random.randint(200,256)
lat_part = ((y2 - y1)*a/256 + y1 )
y = np.array( lat_part + np.mean(lat_part)/(20)*np.sin(a),dtype = 'int')
```

```python
cp = np.array(Image.open('tf.jpeg').resize((256,256)))
skb = np.array([135,206,235])

for i in range(52):
    cp[:a[i],y[i]:,:] = skb
```
